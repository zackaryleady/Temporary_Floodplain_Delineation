```python
import numpy as np
import rasterio
import matplotlib
import matplotlib.pyplot as plt
import skimage
from skimage.filters import (try_all_threshold, threshold_niblack, threshold_sauvola,
                             threshold_mean, threshold_li, threshold_yen, threshold_isodata,
                             threshold_minimum, threshold_triangle, threshold_otsu)
```

```python
# NDWI is a remote sensing index or derived band from other bands that is particularly good at detecting water
def gen_NDWI(image_path):
    with rasterio.open(image_path) as ds:
        green_band2 = ds.read(2)
        nir_band4 = ds.read(4)
        np.seterr(divide='ignore', invalid='ignore')
        ndwi_array = ((green_band2.astype(float) - nir_band4.astype(float)) / (green_band2.astype(float) + nir_band4.astype(float)))
        ds.close()
    return ndwi_array 
```

```python
# load images from data folder
dry_img_path = "./intro_data/reflectance/20161011_final_mosaic.tif"
wet_img_path = "./intro_data/reflectance/20170228_final_mosaic.tif"
ndwi_dry_img = gen_NDWI(dry_img_path)
ndwi_wet_img = gen_NDWI(wet_img_path)
print('The 2D shape of the Dry Image for NDWI on 2016-10-11: {}'.format(ndwi_dry_img.shape))
print('The 2D shape of the Wet Image for NDWI on 2017-02-28: {}'.format(ndwi_wet_img.shape))
print('The shape of the Dry and Wet Image should be equal.')
assert ndwi_dry_img.shape == ndwi_wet_img.shape # what is assert doing?
```

We need to remove "No Data Values". This creates an image for display purposes that maintains the exact shape and creates a modified array for use in histogram analysis. The difference is that the image for display is replacing NaN values as 0, which maintains the array indexing but would effect the data distribution on a histogram, so the histogram array actually removes all the NaN values but changes the array indexing.

```python
# by default nan_to_num is setting No Data Values (NaN) as 0 which should have limited visual impact as NDWI ranges from -1 to 1
ndwi_dry_img_display = np.nan_to_num(ndwi_dry_img)
ndwi_wet_img_display = np.nan_to_num(ndwi_wet_img)
ndwi_dry_histo = ndwi_dry_img[~np.isnan(ndwi_dry_img)]
ndwi_wet_histo = ndwi_wet_img[~np.isnan(ndwi_wet_img)]
assert ndwi_dry_img_display.shape == ndwi_wet_img_display.shape
total_img_pixels = ndwi_dry_img_display.shape[0]*ndwi_dry_img_display.shape[1]
print('Shape of Images for Display: {} | Total number of pixels: {}'.format(ndwi_dry_img_display.shape, total_img_pixels))
print('Number of non-NaN values in the Dry Histogram Array: {} | Removed {} NaN values'.format(ndwi_dry_histo.shape, total_img_pixels - ndwi_dry_histo.shape[0]))
print('Number of non-NaN values in the Wet Histogram Array: {} | Removed {} NaN values'.format(ndwi_wet_histo.shape, total_img_pixels - ndwi_wet_histo.shape[0]))
```

But where are the NaN values we are removing located in our images?

```python
dry_nan = np.where(np.isnan(ndwi_dry_img), 1, 0)
wet_nan = np.where(np.isnan(ndwi_wet_img), 1, 0)
print(dry_nan)
print(wet_nan)
plt.figure(figsize=(12, 12))
plt.subplot(1, 2, 1)
plt.imshow(dry_nan)
plt.title('NDWI Dry Image Location of No Data Values')
plt.subplot(1, 2, 2)
plt.imshow(wet_nan)
plt.title('NDWI Wet Image Location of No Data Values')
plt.show()
```

Let's visualize our NDWI Images and their respective Histograms.

```python
plt.figure(figsize=(24, 16))
plt.subplot(2, 2, 1)
plt.imshow(ndwi_dry_img_display, cmap=plt.cm.inferno)
plt.title('NDWI Dry Image 2016-10-11')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(ndwi_wet_img_display, cmap=plt.cm.inferno)
plt.title('NDWI Wet Image 2017-02-28')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.hist(ndwi_dry_histo, bins=256)
plt.title('Histogram NDWI Dry Image')

plt.subplot(2, 2, 4)
plt.hist(ndwi_wet_histo, bins=256)
plt.title('Histogram NDWI Wet Image')
plt.show()
```

Let's visualize numerous thresholding methods on our two images.<br>The goal at this point is to try to determine which thresholding method works best in seperating water from not water.<br>(Note: the helper function try_all_threshold requires a 2D array, and thus we give it the 2D image array where NaN's are filled as zeros, this has a very small effect on the result.)

```python
fig, ax = try_all_threshold(ndwi_dry_img_display, figsize=(10,8), verbose=True)
plt.show()
```

```python
fig, ax = try_all_threshold(ndwi_wet_img_display, figsize=(10,8), verbose=True)
plt.show()
```

The Minimum technique (threshold_minimum function) is able to accurately separate the water and no water classes from our 1D histogram despite the noisy NaN == 0 inclusion. The other techniques, especially Otsu fail miserably by confusing built-up areas (buildings/roads) with water. This is a well known problem with Otsu thresholding on NDWI.

```python
threshold_dry = threshold_minimum(ndwi_dry_histo)
threshold_wet = threshold_minimum(ndwi_wet_histo)
threshold_dry_otsu = threshold_otsu(ndwi_dry_histo)
threshold_wet_otsu = threshold_otsu(ndwi_wet_histo)
print('Threshold for the Dry Image: {}'.format(threshold_dry))
print('Threshold for the Wet Image: {}'.format(threshold_wet))
ndwi_dry_thresh = ndwi_dry_img_display > threshold_dry
ndwi_wet_thresh = ndwi_wet_img_display > threshold_wet
```

```python
plt.figure(figsize=(24,16))
plt.subplot(2, 2, 1)
plt.imshow(ndwi_dry_thresh, cmap=plt.cm.gray)
plt.title('NDWI Dry Thresholded Image on 2016-10-11')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(ndwi_wet_thresh, cmap=plt.cm.gray)
plt.title('NDWI Wet Thresholded Image 2017-02-28')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.hist(ndwi_dry_histo, bins=256)
plt.axvline(threshold_dry, color='r')
plt.axvline(threshold_dry_otsu, color='black')
plt.title('Histogram of NDWI Dry Image with Thresholds Red=Min Black=Otsu')

plt.subplot(2, 2, 4)
plt.hist(ndwi_wet_histo, bins=256)
plt.axvline(threshold_wet, color='r')
plt.axvline(threshold_dry_otsu, color='black')
plt.title('Histogram of NDWI Dry Image with Thresholds Red=Min Black=Otsu')
plt.show()
```

```python
# ^ is the bitwise boolean operator for subtraction
dry_wet = ndwi_dry_thresh ^ ndwi_wet_thresh
wet_dry = ndwi_wet_thresh ^ ndwi_dry_thresh
assert dry_wet.all() == wet_dry.all()
```

```python
plt.figure(figsize=(24,16))
plt.subplot(1, 2, 1)
plt.imshow(dry_wet, cmap=plt.cm.gray)
plt.title('NDWI Dry-Wet Thresholded Image aka Temporary Floodplain')

plt.subplot(1, 2, 2)
plt.imshow(wet_dry, cmap=plt.cm.gray)
plt.title('NDWI Wet-Dry Thresholded Image aka Temporary Floodplain')
plt.show()
```

```python
wd_sum = np.sum(wet_dry)
dw_sum = np.sum(dry_wet)
assert wd_sum == dw_sum
print('Total number of Temporary Floodplain/Seasonal Wetland pixels: ', wd_sum, dw_sum)
```

```python
# Each pixel is 3.1 meters x 3.1 meters
pixel_area = 3.1*3.1
total_area_true = dw_sum*pixel_area
print('Area of each pixel in square meters: ', pixel_area)
print('Total area of Temporary Floodplain/Seasonal Wetland in square meters: ', total_area_true)
```

```python
# Acres
total_area_true_acres = total_area_true / 4046.856 # constant
print('Total area in acres of Temporary Floodplain/Seasonal Wetland detected: ',
      round(total_area_true_acres, 2))
```

```python
wet_sum = np.sum(ndwi_wet_thresh)
dry_sum = np.sum(ndwi_dry_thresh)
print('Total number of water pixels: '+'Dry Image: {}'.format(dry_sum)+'     '+'Wet Image: {}'.format(wet_sum))
print('Total area in square meters of water pixels: '+'Dry Image: {}'.format(dry_sum*pixel_area)+'     '+'Wet Image: {}'.format(wet_sum*pixel_area))
print('Total area in acres of of water pixels: '+'Dry Image: {}'.format(round((dry_sum*pixel_area)/4046.856, 2))+ '     ' + 'Wet Image: {}'.format(round((wet_sum*pixel_area)/4046.856, 2)))
print('Total area in acres of target == {} or {} - {} = {}'.format(round(total_area_true_acres, 2), 
                                                                   round((wet_sum*pixel_area)/4046.856, 2), 
                                                                   round((dry_sum*pixel_area)/4046.856, 2), 
                                                                   round((wet_sum*pixel_area)/4046.856, 2)-round((dry_sum*pixel_area)/4046.856, 2)))
print('Small differences likely due to rounding errors')
```

```python
dry_wet_sum = ndwi_dry_thresh.astype(int) + ndwi_wet_thresh.astype(int)
wet_dry_sum = ndwi_wet_thresh.astype(int) + ndwi_dry_thresh.astype(int)
print(np.unique(dry_wet_sum))
Customcmap = matplotlib.colors.ListedColormap(['black', 'red', 'blue'])
plt.figure(figsize=(24,16))
plt.subplot(1, 2, 1)
plt.imshow(dry_wet_sum, cmap=Customcmap)
plt.title('NDWI Dry+Wet Thresholded Image Black=No Water, Red=Temporary Water, Blue=Permanent Water')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(wet_dry_sum, cmap=Customcmap)
plt.title('NDWI Wet+Dry Thresholded Image Black=No Water, Red=Temporary Water, Blue=Permanent Water')
plt.axis('off')
plt.show()
```
